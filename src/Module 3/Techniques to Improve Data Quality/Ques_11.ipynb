```python
import pandas as pd

# Step 1: Load the dataset
# Replace with your CSV file path
df = pd.read_csv('/workspaces/AI_DATA_ANALYSIS_/src/Module 3/Techniques to Improve Data Quality/employees.csv')

print("Original Data:")
print(df)

# Task 1: Dropping missing data
# Inspect missing data count per column
print("\nMissing values per column:")
print(df.isnull().sum())

# Drop rows with any missing values
df_dropped = df.dropna()
print("\nData after dropping rows with missing values:")
print(df_dropped)

# Save the dropped dataset
df_dropped.to_csv('employees_dropped.csv', index=False)


# Task 2: Imputation using Mean (for numerical columns)
df_mean_imputed = df.copy()

# Fill missing numerical columns with mean
num_cols = df_mean_imputed.select_dtypes(include=['number']).columns
for col in num_cols:
    mean_val = df_mean_imputed[col].mean()
    df_mean_imputed[col].fillna(mean_val, inplace=True)

print("\nData after mean imputation (numerical columns):")
print(df_mean_imputed)

df_mean_imputed.to_csv('employees_mean_imputed.csv', index=False)


# Task 3: Imputation using Median (numerical) and Mode (categorical)
df_median_mode_imputed = df.copy()

# Fill numerical columns with median
for col in num_cols:
    median_val = df_median_mode_imputed[col].median()
    df_median_mode_imputed[col].fillna(median_val, inplace=True)

# Fill categorical columns with mode
cat_cols = df_median_mode_imputed.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    mode_val = df_median_mode_imputed[col].mode()
    if not mode_val.empty:
        df_median_mode_imputed[col].fillna(mode_val[0], inplace=True)

print("\nData after median (numerical) and mode (categorical) imputation:")
print(df_median_mode_imputed)

df_median_mode_imputed.to_csv('employees_median_mode_imputed.csv', index=False)
```
