```python
# Question: Detecting Data Drift
# Description: Identify potential data drift between two time periods for a numeric attribute.


import pandas as pd
import numpy as np
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/workspaces/AI_DATA_ANALYSIS_/src/Module 3/common_data_errors_example.csv")

# Example: Compare revenue between Janâ€“Mar vs Aprâ€“Jun
period_1 = df[df['month'].isin(['Jan', 'Feb', 'Mar'])]['revenue']
period_2 = df[df['month'].isin(['Apr', 'May', 'Jun'])]['revenue']

# Kolmogorov-Smirnov test (distribution comparison)
ks_stat, ks_pval = ks_2samp(period_1, period_2)

# Output results
print(f"KS Statistic: {ks_stat:.4f}")
print(f"P-value: {ks_pval:.4f}")

if ks_pval < 0.05:
    print("âš ï¸ Significant data drift detected between the two periods.")
else:
    print("âœ… No significant data drift detected.")

# Optional: Visualization
plt.figure(figsize=(10, 6))
sns.kdeplot(period_1, label='Period 1 (Jan-Mar)', fill=True)
sns.kdeplot(period_2, label='Period 2 (Apr-Jun)', fill=True)
plt.title("Revenue Distribution Comparison")
plt.xlabel("Revenue")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
