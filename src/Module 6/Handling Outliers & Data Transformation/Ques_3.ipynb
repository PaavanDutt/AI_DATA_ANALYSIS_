```python
# Question: Power Transformation (Box-Cox) Method
# Description: Use Box-Cox transformation to stabilize variance and make the data more normal.


import numpy as np
from scipy.stats import boxcox
import matplotlib.pyplot as plt
import seaborn as sns

# Create positively skewed data
data = np.random.exponential(scale=2.0, size=1000)

# Apply Box-Cox transformation
data_boxcox, fitted_lambda = boxcox(data)

# Plot comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data, bins=30, kde=True, ax=axes[0])
axes[0].set_title("Original Data")
sns.histplot(data_boxcox, bins=30, kde=True, ax=axes[1])
axes[1].set_title(f"Box-Cox Transformed Data (={fitted_lambda:.2f})")
plt.show()

```
