```python
# Advanced Seaborn Plot with DataFrame
# Description: Create a Seaborn plot using a DataFrame that visualizes two features in a single plot. Use tips
# dataset available in Seaborn's sample datasets.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a Seaborn scatter plot with two features and categorical encoding
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',          # Categorical coloring
    style='smoker',     # Categorical marker style
    palette='Set2',
    s=100               # Marker size
)

# Customize the plot
plt.title('Total Bill vs. Tip by Gender and Smoking Status', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
