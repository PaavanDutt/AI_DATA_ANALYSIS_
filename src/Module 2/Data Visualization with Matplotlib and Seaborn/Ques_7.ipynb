```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Height': [150, 160, 170, 180, 190],
    'Weight': [50, 60, 65, 80, 85],
    'Size': [10, 15, 13, 18, 12]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create scatter plot with size mapping
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(
    data=df,
    x='Height',
    y='Weight',
    size='Size',
    sizes=(50, 300),  # Scale point sizes
    hue='Size',       # Optional: color by size
    palette='viridis',  # Optional: color scheme
    legend='full'
)

# Customize the plot
plt.title('Height vs. Weight with Size-Based Points', fontsize=14, fontweight='bold')
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
