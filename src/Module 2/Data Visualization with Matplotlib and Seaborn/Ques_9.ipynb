```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.


import matplotlib.pyplot as plt

# Data for bar plot
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Data for scatter plot
height = [150, 160, 170, 180, 190]
weight = [50, 60, 65, 80, 85]

# Create subplots: 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Left Plot: Bar Plot for Product Sales ---
axes[0].bar(products, sales, color='mediumseagreen', edgecolor='black')
axes[0].set_title('Product Sales', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Sales (in Thousands)')
axes[0].grid(True, linestyle='--', linewidth=0.5)

# --- Right Plot: Scatter Plot for Height vs. Weight ---
axes[1].scatter(height, weight, color='steelblue', edgecolors='black')
axes[1].set_title('Height vs. Weight', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')
axes[1].grid(True, linestyle='--', linewidth=0.5)

# Adjust layout
plt

```
