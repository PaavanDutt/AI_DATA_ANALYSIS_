```python
# Basic Seaborn Bar Plot
# Description: Using Seaborn, plot the same bar plot from Question 1 to compare the sales data.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data: marks of 10 students
marks = [23, 45, 56, 78, 89, 90, 67, 88, 76, 59]

# Create a simple dataframe for plotting
df = pd.DataFrame({
    'Student': [f'Student {i+1}' for i in range(len(marks))],
    'Marks': marks
})

# Create bar plot
sns.barplot(x='Student', y='Marks', data=df)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add title
plt.title('Marks of Students')

plt.show()
```
