```python
# Scatter Plot of Variable Relationships
# Description: Use Seaborn to create a scatter plot showing the relationship between two
# variables.


import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset('tips')

# Create a basic scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', color='dodgerblue', edgecolor='black')

# Customize the plot
plt.title('Relationship Between Total Bill and Tip', fontsize=14, fontweight='bold')
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
