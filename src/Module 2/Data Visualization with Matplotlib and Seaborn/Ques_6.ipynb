```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.



import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Apply custom font settings
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 12

# Create the figure and axes
fig, ax = plt.subplots()

# Create the customized bar plot
bars = ax.bar(products, sales, color='mediumseagreen', edgecolor='black')

# Add grid lines on the Y-axis
ax.yaxis.grid(True, linestyle='--', linewidth=0.8, color='gray')
ax.set_axisbelow(True)  # Ensure grid is below bars

# Set title and axis labels with custom font styling
ax.set_title('Product Sales (in Thousands)', fontsize=16, fontweight='bold')
ax.set_xlabel('Product', fontsize=14)
ax.set_ylabel('Sales (in Thousands)', fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```
