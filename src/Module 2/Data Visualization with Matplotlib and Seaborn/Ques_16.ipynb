```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.


import matplotlib.pyplot as plt

# Sample data for line plot
x = [1, 2, 3, 4, 5]
y_line = [2, 3, 5, 7, 11]

# Sample data for bar plot
categories = ['A', 'B', 'C', 'D', 'E']
y_bar = [5, 7, 3, 8, 4]

# Create subplots: 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot line plot on the first subplot
axes[0].plot(x, y_line, color='blue', marker='o', linestyle='-', linewidth=2)
axes[0].set_title('Line Plot')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].grid(True)

# Plot bar plot on the second subplot
axes[1].bar(categories, y_bar, color='orange')
axes[1].set_title('Bar Plot')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Values')
axes[1].grid(True, axis='y')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
