```python
# Advanced Seaborn Plot - Pairplot
# Objective: Use Seaborn to create a pairplot to visualize pairwise relationships in a dataset.


import seaborn as sns
import matplotlib.pyplot as plt

# Load the sample iris dataset
iris = sns.load_dataset('iris')

# Create a pairplot
sns.pairplot(
    data=iris,
    hue='species',        # Color-code by species
    palette='Set1',       # Color palette
    diag_kind='kde'       # KDE plots on the diagonal
)

# Show the plot
plt.show()
```
