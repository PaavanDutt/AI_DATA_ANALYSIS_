```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Customize fonts globally
plt.rcParams['font.family'] = 'Calibri'  # Or any font available on your system
plt.rcParams['font.size'] = 12

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(
    x, y,
    color='crimson',        # Line color
    linestyle='--',         # Dashed line
    linewidth=2.5,          # Line thickness
    marker='o',             # Marker style
    markerfacecolor='yellow', # Marker fill color
    markeredgewidth=1.5,
    markeredgecolor='black'
)

# Add title and labels with custom font weight
plt.title('Customized Line Plot', fontsize=16, fontweight='bold')
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

# Add grid with custom style
plt.grid(True, linestyle=':', linewidth=1, color='gray')

plt.tight_layout()
plt.show()
```
