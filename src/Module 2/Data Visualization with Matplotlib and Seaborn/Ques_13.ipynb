```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.


import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create a customized boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=iris,
    x='species',
    y='petal_length',
    palette='pastel',       # Use a soft color palette
    width=0.6,              # Width of the boxes
    linewidth=1.5,          # Thickness of box edges
    fliersize=5             # Size of outlier points
)

# Customize plot aesthetics
plt.title('Distribution of Petal Length by Species', fontsize=14, fontweight='bold')
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5, axis='y')

plt.tight_layout()
plt.show()
```
