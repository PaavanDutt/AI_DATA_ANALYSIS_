```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for height (in cm) and weight (in kg)
data = {
    'Height': [150, 160, 170, 180, 190, 155, 165, 175, 185, 195],
    'Weight': [50, 60, 65, 80, 90, 55, 62, 75, 85, 95]
}

df = pd.DataFrame(data)

# Create scatter plot
sns.scatterplot(x='Height', y='Weight', data=df)

# Add title and labels
plt.title('Height vs. Weight Scatter Plot')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

plt.show()

```
