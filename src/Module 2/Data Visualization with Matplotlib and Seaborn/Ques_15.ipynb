```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Select only numeric columns to compute correlation
corr = iris.select_dtypes(include='number').corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5,
    linecolor='gray',
    cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'},
    square=True,
    annot_kws={"size": 12, "weight": "bold", "color": "black"}
)

plt.title('Correlation Matrix Heatmap - Iris Dataset', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
