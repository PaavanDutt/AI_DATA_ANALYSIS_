```python
# Title: Feature Engineering for Classification
# Description: Create a new feature that could help distinguish between species based on
# logical assumptions and verify its utility.

import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Create a new feature: petal length to petal width ratio
iris['petal_ratio'] = iris['petal_length'] / iris['petal_width']

# Display basic statistics of the new feature grouped by species
print(iris.groupby('species')['petal_ratio'].describe())

# Visualize the distribution of the new feature by species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_ratio', data=iris, palette='Set2')
plt.title('Distribution of Petal Length to Width Ratio by Species')
plt.ylabel('Petal Length / Petal Width Ratio')
plt.xlabel('Species')
plt.show()
```
