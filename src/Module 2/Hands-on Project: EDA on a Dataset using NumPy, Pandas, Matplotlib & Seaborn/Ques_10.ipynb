```python
# 3. Load a CSV file into a DataFrame
# Replace 'your_data.csv' with your actual CSV file path
df = pd.read_csv('/workspaces/AI_DATA_ANALYSIS_/src/Module 2/swiggy.csv')

# 4. Display the first few rows
print(df.head())

# 5. Get a summary of the dataset
print(df.info())
print(df.describe())
```

```python
# 6. Convert DataFrame column to NumPy array and perform operations
col_array = df['Price'].to_numpy()
print("Mean:", np.mean(col_array))
print("Sum:", np.sum(col_array))

# 7. Create a NumPy array and calculate variance and std deviation
arr = np.array([10, 20, 30, 40, 50])
print("Variance:", np.var(arr))
print("Standard Deviation:", np.std(arr))

# 8. Filter NumPy array based on condition (e.g., values > 25)
filtered_arr = arr[arr > 25]
print("Filtered array (values > 25):", filtered_arr)
```

```python
# Fill numeric columns with median
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical columns with mode
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# 12. Basic plot with Matplotlib
plt.plot(df['Price'])
plt.title('Basic Line Plot')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()

# 13. Histogram using Seaborn
sns.histplot(df['Price'], bins=20, kde=True)
plt.title('Histogram with KDE')
plt.show()

# 14. Boxplot with Seaborn
sns.boxplot(x='Area', y='Price', data=df)
plt.title('Boxplot by Category')
plt.show()
```
