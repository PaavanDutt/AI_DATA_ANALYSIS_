```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Plot the distribution of sepal_length
plt.figure(figsize=(8, 5))
sns.histplot(iris['sepal_length'], kde=True, color='skyblue', bins=15)

plt.title('Distribution of Sepal Length', fontsize=16, fontweight='bold')
plt.xlabel('Sepal Length (cm)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)


plt.tight_layout()
plt.show()
```
