```python
import pandas as pd

# 1. Load the Data
# For example, loading a CSV file
df = pd.read_csv('/workspaces/AI_DATA_ANALYSIS_/src/Module 3/common_data_errors_example.csv')

# 2. Detect Missing Data
# Show the count of missing values per column
missing_counts = df.isnull().sum()
print("Missing values per column:")
print(missing_counts)

# Or get a boolean dataframe showing missing values
missing_bool = df.isnull()
print("Boolean mask of missing data:")
print(missing_bool)
```

```python
# 1. Drop Rows with Missing Values
df_dropped = df.dropna()

print("Data after dropping rows with missing values:")
print(df_dropped)
```

```python
# 1. Fill Missing Values
# Example: Fill missing values with the column mean for numeric columns

# Select only numeric columns for mean imputation
numeric_cols = df.select_dtypes(include=['number']).columns

# Fill missing numeric columns with their mean
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# For categorical columns, fill with mode or a placeholder, example:
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_cols:
    mode_val = df[col].mode()
    if not mode_val.empty:
        df[col] = df[col].fillna(mode_val[0])
    else:
        df[col] = df[col].fillna('Unknown')  # fallback option

print("Data after imputing missing values:")
print(df)


# For categorical columns, you might fill with mode or a placeholder string:
# df['category_column'] = df['category_column'].fillna(df['category_column'].mode()[0])

```
